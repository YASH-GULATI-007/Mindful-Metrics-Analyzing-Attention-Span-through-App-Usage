```python
# Import Statements

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading Data

df = pd.read_csv("prepared_data.csv")
df.head()
```

```python
# Count Plot

sns.countplot(data=df, x='attention_span')
plt.title('Distribution of Attention Span')
plt.xlabel('Attention Span (Encoded)')
plt.ylabel('Count')
plt.show()
```

```python
# From the above we can see the count of Attention Span.
# The 1 shows us that 10 - 30 minutes is the most common range
```

```python
# Heat Map

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Models like SVR or Gradient Boosting are best for these
# The Dataset has complex, weak and nonlinear realtionships
```

```python
# Average Screen Time Compared to Attention Span

sns.boxplot(data=df, x='attention_span', y='average_screen_time')
plt.title('Screen Time vs Attention Span')
plt.xlabel('Attention Span')
plt.ylabel('Average Screen Time')
plt.show()
```

```python
# Screen Time is not a sole Feature that effects the Attention Span
df = df.drop('average_screen_time', axis = 1)
df.head()
```

```python
# App Usage Compared to Attention Span

sns.boxplot(data=df, x='app_group_Utility', y='attention_span')
plt.xticks([0, 1], ['Leisure', 'Utility'])
plt.title('Attention Span by App Category')
plt.xlabel('App Group')
plt.ylabel('Attention Span')
plt.show()
```

```python
# People with use of Liesure apps are more likely to have lesser Attention Span
```

```python
# Use of Productivity App Compared to Attention Span

sns.boxplot(data=df, x='uses_productivity_apps', y='attention_span')
plt.title('Uses Productivity Apps vs Attention Span')
plt.xticks([0, 1,2], ['Yes', 'Yes,but no help','No'])
plt.ylabel('Attention Span')
plt.xlabel('Productivity Apps')
plt.show()
```

```python
# People that dont use app generally have more attention span and the group with yes are same on average
df['uses_productivity_apps'] = df['uses_productivity_apps'].map({
    0 : 0,
    1 : 0, 
    2 : 1 
})
df.head()
```

```python
# Notification Responses effects on Attention Span

sns.boxplot(data=df, x='notification_response', y='attention_span')
plt.title('Notification Responses effects on Attention Span')
plt.ylabel('Attention Span')
plt.xlabel('Notification Response')
plt.show()
```

```python
# Shows some effect so we can let it bei in the final evaluation dataset
```

```python
# Screen ACtivity relation to Attention Span

sns.boxplot(data=df, x='screen_activity_numeric', y='attention_span')
plt.title('Screen Activity reltion to Attention Span')
plt.xticks([0, 1], ['Entertainment', 'Academic'])
plt.ylabel('Attention Span')
plt.xlabel('Screen Activity')
plt.show()
```

```python
# This group alse influences the data so we can use it to make final predictions
```

```python
# The data is now ready for final sets lets get it out as a CSV file
df.to_csv("processed_data.csv", index=False)
```
